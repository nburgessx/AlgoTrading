## Socket Communication

_**This requires the tick data server to run separately (different Python kernel).**_

```python
import zmq
import datetime as dt

context = zmq.Context()
socket = context.socket(zmq.SUB)
socket.connect('tcp://127.0.0.1:5555')
socket.setsockopt_string(zmq.SUBSCRIBE, '')
```

## Plotting Canvas

```python
import plotly.graph_objects as go
```

```python
fig = go.FigureWidget()
fig.add_scatter()
fig.add_scatter()
fig
```

## Streaming Plot

```python
import json
```

```python
times = list()
aapl = [None]
msft = [None]
while True:
    msg = socket.recv_string()
    data = json.loads(msg)
    t = dt.datetime.now()
    times.append(t)
    symbol, price = data['SYMBOL'], data['PRICE']
    if symbol == 'MSFT':
        msft.append(float(price))
        aapl.append(aapl[-1])
        fig.data[0].x = times
        fig.data[0].y = msft
    if symbol == 'AAPL':
        aapl.append(float(price))
        msft.append(msft[-1])
        fig.data[1].x = times
        fig.data[1].y = aapl
```
