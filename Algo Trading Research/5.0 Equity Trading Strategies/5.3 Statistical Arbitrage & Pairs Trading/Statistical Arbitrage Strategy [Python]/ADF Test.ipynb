## Read CSV file

```python
# To read csv file
import pandas as pd

# Allow Cell Multi Printing
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

# Disable Warnings
import warnings
warnings.filterwarnings('ignore')

# Read CSV into dataframe data
data = pd.read_csv('AUDCAD.csv',index_col=0)
data.index = pd.to_datetime(data.index)
data.head()
```

## Plot the price series

```python
# Import matplotlib to plot the graph
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.figsize'] = (10,7)

# Plot the prices series
data.Close.plot(color='b')
plt.ylabel('Price')

# Show the plot
plt.show()
```

## Perform ADF Test

```python
# To perform ADF Test
from statsmodels.tsa.stattools import adfuller

# To get help
#?adfuller

# or
#help(adfuller)
```

```python
# Compute ADF test statistics
adf = adfuller(data.Close, maxlag = 1)
```

## AD Fuller Results

```python
print("ADF test results:")
print(f"Test statistic: {adf[0]:.2f}")
print(f"P-value: {adf[1]:.2f}")
print(f"Number of lags used: {adf[2]:.2f}")
print(f"Critical values: {adf[4]}")
```
